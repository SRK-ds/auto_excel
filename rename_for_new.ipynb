```python
import os
from openpyxl import load_workbook
import re

# 특수 문자를 제거하는 함수
def sanitize_filename(value):
    # 파일 이름에 사용할 수 없는 문자 목록: <>:"/\|?*
    return re.sub(r'[<>:"/\\|?*]', '_', str(value))

# 엑셀 파일이 저장된 폴더 경로
folder_path = r'C:\Users\NT551\Desktop\test\test'

# 폴더 내의 모든 엑셀 파일을 처리
for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file_name)
        
        try:
            # 엑셀 파일 열기 (data_only=True로 수식 결과를 가져옴)
            workbook = load_workbook(file_path, data_only=True)

            # 시트 이름을 모를 때, 첫 번째와 두 번째 시트로 접근
            sheet1 = workbook[workbook.sheetnames[0]]  # 첫 번째 시트
            sheet2 = workbook[workbook.sheetnames[1]]  # 두 번째 시트

            # 시트에서 필요한 셀 값 읽기 (값이 없을 때 기본값 설정)
            value_c24 = sheet2['C24'].value or 'C24_empty'
            value_b5 = sheet1['B5'].value or 'B5_empty'
            value_b3 = sheet1['B3'].value or 'B3_empty'
            value_b1 = sheet2['B1'].value or 'B1_empty'
            value_a5 = sheet1['A5'].value or 'A5_empty'

            # 파일명을 C24_A5_B3_B1.xlsx 형태로 변경
            sanitized_c24 = sanitize_filename(value_c24)
            sanitized_b5 = sanitize_filename(value_b5)
            sanitized_b3 = sanitize_filename(value_b3)
            sanitized_a3 = sanitize_filename(value_a5)

            new_file_name = f"{sanitized_b3}_{sanitized_c24}_{sanitized_a3}_{sanitized_b5}.xlsx"
            new_file_path = os.path.join(folder_path, new_file_name)
            
            # 파일 이름 변경
            os.rename(file_path, new_file_path)
            print(f"Renamed {file_name} to {new_file_name}")
        
        except Exception as e:
            print(f"An error occurred with file {file_name}: {e}")
```
